```python
learning_rate = 0.001
n_epochs = 50

# Weight initialization
w = np.zeros(5)     # four for features and one for bias

for e in range(n_epochs):
    acc_epoch = 0        #accuracy at each epoch
    loss_epoch = 0       #loss at each epoch
    
    for i in range(ds['data'].shape[0]):         #ds['data'].shape[0] is total number of rows in dataset ie. 150
        x = ds['data'][i, :]                     #ds['data'][i, :] is ith row and all columns
        y = int(ds['target'][i] != 0)            #ds['target'][i] is target value at ith row

        # Regression / Hypothesis equation
        z = w[0] + np.sum(x * w[1:])
        
        # Sigmoid Equation
        p = 1/(1+np.exp(-z))
        
        # make a decision
        y_pred = int(p >= 0.5)
        
        # update accuracy
        if y_pred == y:
            acc_epoch += 1
            
        # calculate and update cross-entropy loss value
        loss_epoch += -(y * np.log(p) + (1-y) * np.log(1-p))  
        
        # update weights of the model
        w[0] -= learning_rate * (p-y) * 1
        
        for j in range(x.shape[0]):
            w[j + 1] -= learning_rate * (p-y) * x[j]
        
    # average accuracy and loss
    acc_epoch /= ds['data'].shape[0]
    loss_epoch /= ds['data'].shape[0]
    
    print("At epoch " + str(e) + ':' + ' Accuracy : ' + str(acc_epoch) + " "*10 + 'Loss : ' + str(loss_epoch) + '\n')
```
